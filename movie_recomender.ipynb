```python
def cnt(s):
    l=[]
    c=0
    for i in literal_eval(s):
        if c!=3:
            l.append(i['name'])
            c+=1
        else:
            break
    return l   
```

```python
def cno(s):
    l=[]
    c=0
    for i in literal_eval(s):
        l.append(i['name'])
    return l  
```

```python
def cnp(s):
    l=[]
    c=0
    for i in literal_eval(s):
        if i['job']=="Director":
            l.append(i['name']) 
    return l  
```

```python
def stm(v):
    l=[]
    for i in v.split():
        l.append(ps.stem(i))
    return " ".join(l)
        
```

```python
def recomender(mvi):
    mi=dff[dff['title']==mvi].index[0]
    d=m[mi]
    ml=sorted(list(enumerate(d)),reverse=True,key=lambda x: x[1])[1:6]
    
    for i in ml:
        print(dff.iloc[i[0]].title)
    
```
